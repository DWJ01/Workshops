## Let's compare search algorithms by the time it takes to complete the search.

Find the indices of the five listings where "ocean_proximity" is "ISLAND"

### Method 1: using for loops.

````python
%%time

island_indices = []

for i in range(len(data['ocean_proximity'])):
    if data['ocean_proximity'][i] == 'ISLAND':
        island_indices.append(i)
        
for index in island_indices:
    print(index, data.as_matrix()[index])
````

### Method 2: using pandas

````python
%%time

island_lots = data[data['ocean_proximity'] == "ISLAND"]

print(island_lots)
````

## Let's write a search function to be able to search the rows between two values.

````python
def search_function(lower, upper, column, data):
    first_cond = data[data[str(column)] > lower]
    combined = first_cond[first_cond[str(column)] < upper]
    return combined
````

````python
%%time

narrowed_df = search_function(26, 30, "housing_median_age", data)
````

## Compare it with the typical if-else/for loop method

````python
def slow_search_function(lower, upper, column, data):
    indices = []
    
    for i in range(len(data)):
        if data[str(column)][i] > lower and data[str(column)][i] < upper:
            indices.append(i)
            
    return indices
            
````

````python
%%time

slow_search_function(26, 30, "housing_median_age", data)
````
